```python
import pandas as _pd

def getBdsimTrackInDF(inputfilename):

    data = _bd.Data.Load(inputfilename)
    e = data.GetEvent()
    et = data.GetEventTree()
    npart = et.GetEntries()

    sampler_names = e.GetSamplerNames()
    data_dict = {'S': [], 'X': [], 'Y': [], 'PX': [], 'PY': []}
    sampler_list = []
    for particle in range(npart):
        et.GetEntry(particle)
        for i_sampler, sampler in enumerate(sampler_names):
            if particle == 0 :
                sampler_name = str(sampler).replace('.', '')
                sampler_list.append(sampler_name)
            sampler_data = e.GetSampler(sampler)
            data_dict['S'].append(sampler_data.S)
            data_dict['X'].append(sampler_data.x[0])
            data_dict['Y'].append(sampler_data.y[0])
            data_dict['PX'].append(sampler_data.xp[0])
            data_dict['PY'].append(sampler_data.yp[0])
    for key in data_dict:
        data_dict[key] = _np.asarray(data_dict[key]).flatten()
    df = _pd.DataFrame(data_dict, index=_pd.MultiIndex.from_product([range(s) for s in (npart, len(sampler_names))], 
                                                                    names=['PARTICLE', 'SAMPLER']))
    df.index.set_levels([sampler_list], level=[1], inplace=True)
    return df
```

```python
import time
start_time = time.time()
df = getBdsimTrackInDF(bdsimfile1000_0)
end_time = time.time()
print(end_time - start_time)
```
