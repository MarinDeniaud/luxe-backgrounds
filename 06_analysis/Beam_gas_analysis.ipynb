```python
import numpy as _np
import matplotlib.pyplot as _plt
import glob as _gl
import pymad8 as _m8
import pybdsim as _bd
import beamGas

date = "2023_03_07"
save_name_prefix = "plots/BeamGas/" + date + "_T20_plot"
```

```python
filelist = ['T20_bias_0.200_merged_hist.root', 'T20_bias_0.500_merged_hist.root',
            'T20_bias_1.000_merged_hist.root', 'T20_bias_2.000_merged_hist.root']

twiss = _m8.Output("../01_mad8/TWISS_CL_T20")
```

```python
_plt.rcParams['font.size'] = 20
fig, ax = _plt.subplots(1,1,figsize=(18,5))

beamGas.plot_multiple_hist(filelist, ["h_PFH_S"], linFit=True, fitRange=[0,-1], yLogScale=False)
_plt.xlabel("S [m]")
_plt.ylabel("$N_{PFH}$")
_plt.legend(prop={'size': 10})

_m8.Plot.AddMachineLatticeToFigure(_plt.gcf(), twiss)
#_plt.title('Primary First Hits all processes')
_plt.savefig("{}_Primary_First_Hits_for_all_processes".format(save_name_prefix), dpi=300)
```

```python
_plt.rcParams['font.size'] = 25
fig, ax = _plt.subplots(1,1,figsize=(18,7))

beamGas.plot_multiple_hist(filelist, ["h_PFH_S_unweight"], expFit=False, fitRange=[0,-1], yLogScale=True, printLegend=False)
_plt.xlabel("$s$ [m]")
_plt.ylabel("$N_{\\rm PFH}$")
_plt.legend(['$\chi_{\\rm Total}$ = 0.2','$\chi_{\\rm Total}$ = 0.5','$\chi_{\\rm Total}$ = 1','$\chi_{\\rm Total}$ = 2'], loc=1)

_plt.ylim(2e6,6e8)
_m8.Plot.AddMachineLatticeToFigure(_plt.gcf(), twiss)
_plt.xlim(0,290)
#_plt.title('Primary First Hits all processes (unweighted)')
_plt.savefig("{}_Primary_First_Hits_for_all_processes_unweighted".format(save_name_prefix), dpi=300)
```

```python
_plt.rcParams['font.size'] = 22
fig, ax = _plt.subplots(1,1,figsize=(10,7))

beamGas.plot_var("localhistlist", "h_PFH_S", errorBars=False, fit=False, xLogScale=True, color='C0', printLegend=True)
beamGas.plot_var("farmhistlist", "h_PFH_S", errorBars=False, fit=False, xLogScale=True, color='C1', printLegend=True)
_plt.ylabel("$V_{\\rm PFH}$ [%]")
_plt.xlabel("$\chi_{\\rm Total}$")
_plt.legend(loc=9)
_plt.ylim(13,70)

#_plt.title("Variance")
_plt.savefig("{}_Variance".format(save_name_prefix), dpi=300)
```

```python
# Error on the "Variance" with the 0.5 biasing factor
A = [16.24,16.45,17.31,15.06,16.16,16.29,16.53,16.20]
print("Mean : ", _np.mean(A))
print("Std : ", _np.std(A))
```
