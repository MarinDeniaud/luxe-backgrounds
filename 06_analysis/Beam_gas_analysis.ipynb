```python
import numpy as _np
import matplotlib.pyplot as _plt
import glob as _gl
import pymad8 as _m8
import pybdsim as _bd
import beamGas

date = "2022_11_04"
save_name_prefix = "plots/" + date + "_T20_plot"
```

```python
filelist = ['T20_bias_0.200_merged_hist.root', 'T20_bias_0.500_merged_hist.root', 
            'T20_bias_1.000_merged_hist.root', 'T20_bias_2.000_merged_hist.root']

_plt.figure(figsize=(20, 7))
_plt.subplot(1,2,1)
beamGas.plot_multiple_hist(filelist, ["h_PFH_S"], linFit=True, fitRange=[0,-1], yLogScale=True)
_plt.xlabel("S [m]")
_plt.ylabel("Number of events")
_plt.title('Primary First Hits all processes')

_plt.subplot(1,2,2)
beamGas.plot_multiple_hist(filelist, ["h_PFH_S_unweight"], expFit=True, fitRange=[0,-1], yLogScale=True)
_plt.xlabel("S [m]")
_plt.ylabel("Number of events")
_plt.title('Primary First Hits all processes (unweighted)')
_plt.savefig("{}_Primary_First_Hits_for_all_processes".format(save_name_prefix), dpi=300)
```

```python
_plt.figure(figsize=(10, 7))
beamGas.plot_var("localhistlist", "h_PFH_S", fit=False, xLogScale=True, color='C0', printLegend=True)
beamGas.plot_var("farmhistlist", "h_PFH_S", fit=False, xLogScale=True, color='C1', printLegend=True)
_plt.ylabel("%")
_plt.xlabel("Biasing factor")
_plt.title("Variance")
_plt.savefig("{}_Variance".format(save_name_prefix), dpi=300)
```

```python
# Error on the "Variance" with the 0.5 biasing factor
A = [16.24,16.45,17.31,15.06,16.16,16.29,16.53,16.20]
print("Mean : ", _np.mean(A))
print("Std : ", _np.std(A))
```
